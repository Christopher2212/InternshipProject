```python
import pandas as pd
import networkx as nx
import numpy as np

#Getting a mapping matrix that has the corresponding weight between each node
def getMappingMatrix(nodeListName,edgeListName):
    edgeList = pd.read_csv(edgeListName)
    nodeList = pd.read_csv(nodeListName)
    
    #init the matrix
    mapMat = []
    count = nodeList[nodeList.columns[0]].count()

    for i in range(count):
        temp = []
        for j in range(count):
            temp.append(0)
        mapMat.append(temp)
    
    
    for row in edgeList.iterrows():
        
        lat1 = (nodeList.iloc[row[1]["from"]-1]["latitude"])/(180/np.pi) 
        lat2 = (nodeList.iloc[row[1]["to"]-1]["latitude"])/(180/np.pi)
        long1 = (nodeList.iloc[row[1]["from"]-1]["longitude"])/(180/np.pi)
        long2 = (nodeList.iloc[row[1]["to"]-1]["longitude"])/(180/np.pi)
    
        #Confirmed to be in miles and accurate
        dist = 3963.0 * np.arccos(np.sin(lat1) * np.sin(lat2) + np.cos(lat1) * np.cos(lat2) * np.cos(long2 - long1))
        
        mapMat[row[1]["to"]-1][row[1]["from"]-1] = dist
        
    return mapMat

map = getMappingMatrix("Other\EdgeandVertListsCSV\BangkokVert.csv","Other/EdgeandVertListsCSV/BangkokEdge.csv")
#Have to convert our regular matrix to a numpy matrix TODO: THIS IS HORRIBLE WASTE OF REASOURCE, PLZ CHANGE LATER!
npM = np.matrix(map)
#makes networkxs graph 
G = nx.from_numpy_matrix(npM)
print(G)

#TODO: MAKE SURE CALCULATIONS ARE CORRECT!!!! 
#Centrality calculations!!!!!
centrality = nx.betweenness_centrality(G, k=35,endpoints=True)
print("Betweeness Centrality of each node")
print(centrality)

#returns us the avg clustering of our metro station
#TODO: Avg Clustering is = 0 debug
print("Avg Clustering")
print(nx.average_clustering(G))

print("Density")
print(nx.density(G))
```
